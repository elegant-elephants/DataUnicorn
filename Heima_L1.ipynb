## Action1：求2+4+6+8+...+100的求和，用Python该如何写

```python
sum = 0
a = 2
for a in range(0,99,2):
    sum += a
    a = a+2
    print(a,sum)
print('Action1求和为:',sum)
```

## Step1，数据加载
## Step2，数据预处理
拆分problem类型 => 多个字段
## Step3，数据统计
对数据进行探索：品牌投诉总数，车型投诉总数
哪个品牌的平均车型投诉最多`

```python
df=df.drop('problem', axis=1).join(df.problem.str.get_dummies(','))
print(df)
```

```python
df.head()
```

```python
#数据清洗
def f(x):
    x = x.replace('一汽-大众','一汽大众')
    return x


df['brand'] = df['brand'].apply(f)

result = df.groupby(['brand'])['id'].agg(['count'])
```

```python
result
```

### 问题标记

```python

tags = df.columns[7:]
tags
```

```python
result2 = df.groupby(['brand'])[tags].agg(['sum'])
result2
```
